```python
import pandas as pd
df = pd.read_csv('data.csv',index_col = 'sn')
df[(df['add'] == 'Kathmandu') & (df['age']>30)]           # & - OR 
```

```python
import pandas as pd
df = pd.read_csv('data.csv',index_col = 'sn')
df[(df['add'] == 'Kathmandu') | (df['age']>50)]            # | - OR  
```

```python
import pandas as pd
df = pd.read_csv('data.csv',index_col = 'sn')
df = df[(df['add'] == 'Kathmandu') | (df['age'] > 50)]
df.to_csv('new_data.csv')          #Creates new CSV file
df
```
